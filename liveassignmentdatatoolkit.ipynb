#difference in deep copy and shallow copy
#deep copy>> copies all nested objects,creating entirely independent objects
#shallow copy>> copies the outer object, but the inner objects are shared between the original and the copy

```python
#shallow copy
import copy
lis=[[1,2],[3,4]]
shallow_copy=copy.copy(lis)
shallow_copy[0][0]=10
```

```python
print(lis)
```

```python
print(shallow_copy)
```

```python
#Deep copy
import copy
lis1=[[1,2],[3,4]]
deep_copy=copy.deepcopy(lis1)
deep_copy[0][0]=20
```

```python
print(lis1)
```

```python
print(deep_copy)
```

#Q14 create a time series data in a pandas data frame with columns: date temperature humidity and perform the following task using matplotlib.
#plot temparature and humidity on the same plot wit different y-axis (left y axis for temperature and right y-axis for humidity)
#label the x-axis ae date
#set the title of the plot as 'temperature and humidity overtime'

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dates=pd.date_range(start='1/7/2024',periods=20)
```

```python
temperature=np.random.uniform(25,40,size=20)
humidity=np.random.uniform(20,60,size=20)
```

```python
df=pd.DataFrame({'Date':dates,'Temperature':temperature,'Humidity':humidity})
```

```python
fig,ax1=plt.subplots()

ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature',color='tab:red')
ax1.plot(df['Date'], df['Temperature'], color='tab:red', label='Temperature')
ax1.tick_params(axis='y', labelcolor='tab:red')

ax2 = ax1.twinx()
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.plot(df['Date'], df['Humidity'], color='tab:blue', label='Humidity')
ax2.tick_params(axis='y', labelcolor='tab:blue')

plt.title('Temperature and Humidity Overtime')
```

#Q15.create a numpy array data containing 1000 samples from a normal distribution. perform the following task using matplotlib.
#1]ploy a histogram of the data with 30 bins
#2]overlay a line plot representing the normal distribution probablity density function(pdf)
#3]label x-axis as value and the  y-axis as frequency/probablity.
#4]set the title of the plot as 'histogram with pdf overlay'

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

a = np.random.normal(loc=0, scale=1, size=1000)
```

```python
plt.hist(a, bins=30, density=True, alpha=0.6, color='orange', edgecolor='blue')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)

pdf = norm.pdf(x, loc=0, scale=1)
plt.plot(x, pdf, 'k', linewidth=2)

plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with the PDF Overlay')
```

#Q17Create a seaborn scatter plot of two random arrays, color points based on their position relative to the origin(quadrants) add a legend, label the axis and set the title as quadrant wise scatter plot.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.random.randn(10)
y = np.random.randn(10)
```

```python
x
```

```python
y
```

```python
def quad(x, y):
    if x > 0 and y > 0:
        return 'Quadrant 1'
    elif x < 0 and y > 0:
        return 'Quadrant 2'
    elif x < 0 and y < 0:
        return 'Quadrant 3'
    elif x > 0 and y < 0:
        return 'Quadrant 4'
    else:
        return 'Origin'
```

```python
Q = np.array([quad(xi, yi) for xi, yi in zip(x, y)])
Q
```

```python
df = pd.DataFrame({'x': x,'y': y,'Quadrant': quad})
df
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='tab10')
plt.legend(title='Quadrants')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant Wise Scatter Plot')
```

#Q18 with bokeh ,plot a line chart of sine wave function, add grid line,label the axis, set the title as sine wave function.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

x = np.linspace(0, 4 * np.pi, 500)
y = np.sin(x)

a = figure(title="Sine Wave Function", x_axis_label='x', y_axis_label='sin(x)',
           tools="pan,box_zoom,reset,save")

a.line(x, y, legend_label="sin(x)", line_width=2)

a.grid.grid_line_color = "gray"
a.grid.grid_line_alpha = 0.5

show(a)
```

#Q20 using plotly create a basic line plot of randomly generated data set, label the axis and set the title as simple line plot.

```python
import plotly.graph_objects as go
import numpy as np

x = np.arange(10)
y = np.random.randn(10)

fig = go.Figure()

fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data'))

fig.update_layout( title='Simple Line Plot',xaxis_title='X Axis',
    yaxis_title='Y Axis')
fig.show()
```

#Q21 using plotly create an interactive pie chart of randomly generated data, add label and percentages, and set the title as 'Interactive Pie Chart'.

```python
import plotly.graph_objects as go
import numpy as np

labels = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(10,20, size=len(labels))

fig = go.Figure(data=[go.Pie(labels=labels, values=values,
                             textinfo='label+percent',
                             hole=0.3)])

fig.update_layout(title='Interactive Pie Chart')

fig.show()
```
